````python
path_train = " american-sign-language-dataset/asl/"
classes = [c for c in os.listdir(path_train) if not c.startswith(".") and len(c)==1]
classes.sort()
print(classes)
````

````python
batch_size = 64
imageSize = 64
target_dims = (imageSize, imageSize, 3)
num_classes = 36
train_len = 25291
````

###### [1] This code block initializes a dictionary called d with two keys: "img" and "class".

###### [2] The for loop iterates through each class (i.e., letter or number) in the dataset.

######          - For each class, the inner for loop iterates through each image file in the directory corresponding to that class.

######          - For each image file, the code reads in the image using cv2.imread(), resizes it to the desired dimensions using skimage.transform.resize(), and converts it to a numpy array using np.asarray().

###### [3] The resulting numpy array is then reshaped to have dimensions (-1, imageSize, imageSize, 3), where -1 indicates that the size of the first dimension should be inferred based on the size of the other dimensions (i.e., the number of images in the class).

###### [4] Finally, the numpy array is added to the "img" key in the d dictionary, and the class label (c) is added to the "class" key in the d dictionary.

###### [5] After this loop completes, the d dictionary should contain two lists: "img", which contains the numpy arrays representing the images in the dataset, and "class", which contains the corresponding class labels for each image.

````python
d = {"img" : [], "class" : []}
for c in classes:
    for img in os.listdir(os.path.join(path_train,c)):
        img_file = skimage.transform.resize(cv2.imread(os.path.join(path_train,c,img)), (imageSize, imageSize, 3))
        img_arr = np.asarray(img_file).reshape((-1, imageSize, imageSize, 3))
        d["img"].append(img_arr)
        d["class"].append(c)
````

````python
df = pd.DataFrame(d)
ax = sns.countplot(data=df,x="class")
ax.set(title="Classes distribution")
print("Total number of training data :",len(df))
````

### 5. Modeling <a id=11></a>

<h3 style="text-align:center">convolutional neural network (CNN) model using the Keras</h3>

###### [1] The Sequential() function call initializes a sequential model, which allows the user to add layers to the model in a linear stack. The various layers of the model are added using the add() method.

###### [2] The first layer added to the model is a Conv2D layer, which performs a 2-dimensional convolution operation on the input image. The layer has 32 filters, each with a size of 5x5 pixels. The input_shape parameter specifies the dimensions of the input image, which is 64x64 pixels with 3 color channels (RGB).

###### [3] The next three layers added are Activation layers, which apply a nonlinear activation function (in this case, the Rectified Linear Unit or ReLU function) to the output of the previous layer.

###### [4] The next three layers added are MaxPooling2D layers, which perform 2-dimensional max pooling operations on the output of the previous layer. These layers downsample the input by taking the maximum value of each non-overlapping 2x2 pixel region.

###### [5] The Flatten layer converts the output of the previous layer into a 1-dimensional array, which can be input into a fully connected neural network layer.

###### [6] The next layer added is a fully connected Dense layer with 128 neurons and a ReLU activation function.

###### [7] The final layer added is another Dense layer with num_classes neurons and a softmax activation function, which outputs a probability distribution over the different classes in the classification problem.

###### [8] Finally, the summary() method is called on the model to print out a summary of the layers and parameters in the model, including the number of trainable parameters. This can be useful for debugging and optimizing the model.

````python
model = Sequential()

model.add(Conv2D(32, (5, 5), input_shape=(64, 64, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D((2, 2)))

model.add(Conv2D(64, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D((2, 2)))

model.add(Conv2D(64, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D((2, 2)))

model.add(Flatten())

model.add(Dense(128, activation='relu'))

model.add(Dense(num_classes, activation='softmax'))

model.summary()
````

###### Defines a function make_prediction that takes an image as input and returns a prediction for the image's class using the trained model.

###### The function first loads the input image using the load_img method of the image module and resizes it to a target size of 64x64 pixels. Then, it converts the image to a numpy array using the img_to_array method of the image module.

###### Next, the numpy array is preprocessed to be compatible with the trained model. The expand_dims method is used to add an extra dimension to the numpy array, which represents the batch size. The pixel values of the numpy array are normalized by dividing by 255 and subtracting 0.5.

###### The trained model is then used to make a prediction on the preprocessed numpy array. The prediction is returned as a binary value indicating whether the input image belongs to class 0 or class 1. Finally, the binary prediction is converted to a string representing the predicted class label using the classes list, which contains the class labels in the same order as the binary predictions.

````python
from tensorflow.keras.preprocessing import image
def make_prediction(img):
    img = image.load_img(img,target_size=(64,64))
    img = image.img_to_array(img)
    result = trained_model.predict((np.expand_dims(img,axis=0)/255 - 0.5))[0] > 0.5
    return classes[result.tolist().index(True)]
````

###### The make_prediction function preprocesses the input image, passes it through the trained model, and returns the predicted class label. In `this case, the predicted class label represents the letter "Z" in ASL`.

````python
print(make_prediction("/kaggle/input/asl-dataset/asl_dataset/z/hand1_z_bot_seg_2_cropped.jpeg"))
````
